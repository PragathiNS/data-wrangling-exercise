## Distinct first names in the data set

Two dataframes: namesDF <- name field from df, firstName <- name.firstname field from df
    
Steps:
1. Create the above mentioned dataframes and rename the column name to name.firstname in namesDF
2. Use namesDF.update(firstName) to modifiy the namesDF in place using non-NA values from firstName
3. Drop any nan values from namesDF (These are the original missing names)
4. Use re.match() to match the first part of the name field and remove the name prefixes like Mr., Mrs. or Dr.
       For example: Dr. Morgan Morris is returned as Morgan
5. Use fName['name.firstname'].nunique() to retrieve the number of unique values from fName

```python
namesDF = df['name'].to_frame()
namesDF.rename(columns={'name': 'name.firstname'}, inplace=True)
firstName = df['name.firstname'].to_frame()

namesDF.update(firstName)
namesDF = namesDF.dropna()
namesDF.head(15)
```

```python
# applymap(func) where func matches only the first part of the name field and removes the Prefix of the names
fName = namesDF.applymap(lambda name: re.match(r'^(?:(?:Mr|Mrs|Dr)\.\s)?([\S]+)\s?', str(name)).group(1))
fName.head(15)
```

```python
distNames = fName['name.firstname'].nunique()
print("Number of distinct Names: ", distNames)
```

## Distinct street names in the data set

Two dataframes: addr <- address field from df, addSt <- address.street field from df
    
Steps:
1. Create the above mentioned dataframes 
2. Create addrDF by applying applymap(func) where func strips the first line of the value in the address field and    rename the column name to address.street
2. Use addrDF.update(addSt) to modifiy the addrDF in place using non-NA values from addSt
3. Call regexMatch using applymap() on values of addrDF.
4. regexMatch function
    returns None for those values which starts with PSC|Unit|USNS|USNV|USS|USCGC as these are Postal Service Center     units and others are ship prefixes
    returns the street address for the rest by stripping off the numbers and Apt/Suite number from the values
        For example: 86314 David Pass Apt. 211 is returned as David Pass
5. stName is converted to a set so that the values are distinct and nan and None values are removed from the set.
6. len(stName) is returned which is the length of the distinct Street names in the dataset

```python
addr = df['address'].to_frame()
addSt = df['address.street'].to_frame()

# Split the first line of the address field values
addrDF = addr.applymap(lambda ad: str(ad).split("\n")[0])
addrDF.rename(columns={'address': 'address.street'}, inplace=True)

addrDF.update(addSt)
addrDF.head(15)
```

```python
def regexStMatch(street):
    # returns None when the address starts with PSC|Unit|USNS|USNV|USS|USCGC
    if re.match(r'^(?:PSC|Unit|USNS|USNV|USS|USCGC)', str(street)):
        return
    # returns the streetName from the address field by stripping numbers and Apt/Suite
    return re.match(r'^(?:\d+\s+)?(.*?)(?:\s+(?:Apt\.|Suite)\s+\d+)?$', str(street)).group(1)

stName = addrDF.applymap(regexStMatch)
stName.head(15)
```

```python
# stName is a set so that only distinct values are present
stName = set(stName.to_dict()['address.street'].values())

# Drop those rows with 'nan' and None type
stName.remove('nan')
stName.remove(None)

print("Distinct Street Names: ", len(stName))
```

## 5 most common US area codes in phone number field

US area codes(first 3 digits after the country code)

Treated as US phone numbers if it begins with:
    1. (Country code) 1- or 1 or 1. followed by the regex of 2
        Ex: 1-(420)5089422
            1-420-508-9422
            1(420)5089422
            14205089422
            1.(420)5089422
            1.4205089422\
          
    2. \([1-9]\d\d\)
        Ex: (420)5089422
        
    3. [1-9]\d\d 
        Ex:  4205089422
        
       (2 and 3 kind of values could belong to any Country but for this analysis, its assumed to be US numbers) 

Steps:
1. phoneDF is created from df['phone']
2. Call getAreacode using apply() on values of phoneDF.
4. getAreacode function
    returns the stripped numbers (3 digits which are area codes) by matching the regular expression.
        For ex: 1(420)5089422 returned as 420
                +39(2)5142930962 is passed
                4205089422 returned as 420
5. NaN values are dropped
6. Using value_counts()[:5] on areaCodes we get the 5 most common US area codes

```python
phoneDF = df['phone']

def getAreacode(number):
    try:
        # returns the first 3 digit of a phone number that either starts with 1/1-/(/one of the [1-9]
        return re.match(r'(?:1-|1|1\.)?\(?([1-9]\d\d)\)?', str(number)).group(1)
    except AttributeError:
        pass
# Non-null phone number records

areaCodes = phoneDF.apply(getAreacode)
areaCodes.dropna(inplace=True)
print("5 most common US area codes: ")
areaCodes.value_counts()[:5]
```
